```python
# First plot the nullclines

import numpy as np
import matplotlib.pyplot as plt

def dxdt(x, y, I):
    return x - x**3/3 -y

def dydt(x, y, a):
    return 0.08*(x-0.7-0.8*y)

def x_nullcline(x, a):
    return x - x**3/3

def y_nullcline(y, a):
    return x/0.8-0.7/0.8 

x = np.linspace(-3, 3, 100)
y = np.linspace(-3, 3, 100)

I = 0
a = I
plt.figure(figsize=(10,10), dpi=80)
plt.plot(x, x_nullcline(x, a), label='x nullcline', color = 'red')
plt.plot(y, y_nullcline(y, a), label='y nullcline', color = 'blue')
plt.ylim(-3,3)
plt.legend()
plt.show()
```

```python
#nullclines in a vector field
from scipy.integrate import odeint

xd = np.linspace(-3, 3, 12)
yd = np.linspace(-1, 1, 12)

X, Y = np.meshgrid(xd, yd)

U = dxdt(X, Y, a)
V = dydt(X, Y, a)
plt.figure(figsize=(10,10), dpi =100)
plt.quiver(X, Y, U, V, scale= 20, pivot = "mid" )
plt.plot(x, x_nullcline(x, a), label='x nullcline' , color = 'red')
plt.plot(y, y_nullcline(y, a), label='y nullcline' , color = 'blue')
plt.ylim(-3,3)
plt.xlim(-2,2)

plt.legend()

plt.show()
```
